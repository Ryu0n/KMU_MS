```python
path_week_1 = "./drive/MyDrive/Study/zoom_participants_83169866693.xls"
path_week_2 = "./drive/MyDrive/Study/zoom_participants_86750070616.xls"
```

**학생만 남기고 필터링**

```python
df_1 = pd.read_excel(path_week_1)
df_2 = pd.read_excel(path_week_2)

df_1 = df_1[~(df_1.역할.isin(['교수자']))]
df_2 = df_2[~(df_2.역할.isin(['교수자']))]

display(df_1.head())
display(df_2.head())
```

**필요한 컬럼만 남김**

```python
# 이름 학과 학번 1주차 2주차
key_filter = ['이름', '학번', '참여시작시간']
df_1 = df_1[key_filter]
df_2 = df_2[key_filter]
display(df_1.head())
display(df_2.head())
```

**문자열로부터 시간 필터링**

```python
key_apply, func_apply = '참여시작시간', lambda d: d[:11]
df_1[key_apply] = df_1.참여시작시간.apply(func_apply)
df_2[key_apply] = df_2.참여시작시간.apply(func_apply)
display(df_1.head())
display(df_2.head())
```

**학번으로 중복제거**

```python
df_1 = df_1.drop_duplicates('학번').reset_index()
df_2 = df_2.drop_duplicates('학번').reset_index()
display(df_1.head())
display(df_2.head())
```

**불필요한 컬럼 제거 및 중복되는 컬럼명 수정**

```python
df_1 = df_1.rename(columns={'참여시작시간':'1주차'})
df_2 = df_2.rename(columns={'참여시작시간':'2주차'})
df_1 = df_1.drop(['index'], axis=1)
df_2 = df_2.drop(['index'], axis=1)
display(df_1)
display(df_2)
```

**Outer join**

```python
df_3 = pd.merge(df_1, df_2, how='outer', on=['학번', '이름'])
df_3
```

**학과 삽입**

```python
df_3.insert(1, '학과', ['인공지능전공']*len(df_3))
display(df_3) 
```

**날짜가 존재하는 값들은 O로 대치**

```python
import numpy as np
df_3['1주차'] = df_3['1주차'].apply(lambda s: 'X' if s==np.NaN else 'O')
df_3['2주차'] = df_3['2주차'].apply(lambda s: 'X' if s==np.NaN else 'O')
df_3
```

**엑셀로 내보내기**
