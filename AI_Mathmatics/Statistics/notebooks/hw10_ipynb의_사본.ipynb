Cov(X, Y)  
= E[(X - muX)(Y - muY)]  
= E(XY - XmuY - YmuX + muXmuY)  
= E(XY) - muYE(X) - muXE(Y) + muXmuY  

**E(X) = muX, E(Y) = muY 이므로,,**  

= E(XY) - muYmuX - muXmuY + muXmuY  
= E(XY) - muYmuX  
= E(XY) - E(X)E(Y)

E(XY)  
= 1 × 1 × (2/8) + 2 × 1 × (2/8) + 1 × 2 × (1/8)  
= (1/4) + (1/2) + (1/4)  
= 1  

E(X)  
= 0 × (1/8) + 1 × (3/8) + 2 × (3/8) + 3 × (1/8)  
= (3 + 6 + 3) / 8  
= 12/8

E(Y)  
= 0 × (2/8) + 1 × (4/8) + 2 × (2/8)  
= (4 + 4) / 8  
= 8 / 8  

Cov(X, Y)  
= E(XY) - E(X)E(Y)  
= 1 - 1.5  
= -0.5  

공분산은 각 확률변수 (X, Y)의 절대적인 크기에 영향을 받으므로 단위에 영향으 없앨 필요가 있다.  

**상관계수 = Cov(X, Y) / stdX × stdY**  

E(X^2)  
= 0 × (1/8) + 1^2 × (3/8) + 2^2 × (3/8) + 3^2 × (1/8)  
= (3 + 12 + 9) / 8  
= 24/8

Var(X)  
= E(X^2) - [E(X)]^2  
= (24/8) - (12/8)^2  
= 0.75  

E(Y^2)  
= 0 × (2/8) + 1^2 × (4/8) + 2^2 × (2/8)  
= (4 + 8) / 8  
= 12 / 8  

Var(Y)  
= E(Y^2) - [E(Y)^2]^2  
= (12/8) - (8/8)^2  
= 0.5  

상관계수  
= Cov(X, Y) / stdX × stdY  
= -0.5 / sqrt(0.75) × sqrt(0.5)  
= -0.82
