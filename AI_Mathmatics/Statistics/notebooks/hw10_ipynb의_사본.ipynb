공분산
E(XY) = 2*1*3/8 + 1*2*3/8 = 12/8  
E(X) = 1*3/8 + 2*3/8 + 3*1/8 = 12/8  
E(Y) = 1*3/8 + 2*3/8 + 3*1/8 = 12/8  
Cov(X, Y) = E(XY) - E(X)E(Y) = 12/8 - (12/8)^2 = -0.75  
상관계수  
Var(X) = E(X^2) - {E(X)}^2 = 3  
Var(Y) = E(Y^2) - {E(Y)}^2 = 3  
상관계수 = Conv(X, Y) / (Var(X)*Var(Y)) = -0.75 / 9 = -0.8333
