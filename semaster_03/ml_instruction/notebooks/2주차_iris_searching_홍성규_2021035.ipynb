# Iris 데이터 셋으로부터 원하는 데이터 추출
## Conditions
* setosa 이외의 다른 종
* sepal_length < 6.0
* petal_length < 6.0
* 위와 같은 종들의 sepal_width, petal_width 각각의 평균
* 위와 같은 종들을 sepal_width, petal_width 를 기준으로 내림차순 정렬

```python
other_species = df[~(df.species == 'setosa')]
# species[0] = 'versicolor', species[1] = 'virginica'
species = other_species.species.unique()
other_species
```

```python
under_6 = other_species[(other_species.sepal_length < 6.0) & (other_species.petal_length < 6.0)]
under_6
```

```python
versicolor_6, virginica_6 = under_6[(under_6.species==species[0])], under_6[(under_6.species==species[1])]
```

```python
versicolor_6.head()
```

```python
virginica_6.head()
```

```python
df_new = pd.DataFrame(
    [[versicolor_6.sepal_width.mean(), versicolor_6.petal_width.mean()],
    [virginica_6.sepal_width.mean(), virginica_6.petal_width.mean()]],
    index=species,
    columns=['sepal_width_mean', 'petal_width_mean']
)
df_new
```

```python
sorted_versicolor = versicolor_6.sort_values(by='sepal_width', ascending=False)
sorted_versicolor.head()
```

```python
sorted_versicolor = versicolor_6.sort_values(by='petal_width', ascending=False)
sorted_versicolor.head()
```

```python
sorted_virginica = virginica_6.sort_values(by='sepal_width', ascending=False)
sorted_virginica.head()
```

```python
sorted_virginica = virginica_6.sort_values(by='petal_width', ascending=False)
sorted_virginica.head()
```
